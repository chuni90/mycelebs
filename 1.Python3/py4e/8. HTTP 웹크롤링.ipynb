### urllib 라이브러리
- urllib 라이브러리를 활용하여 간단한 웹 브라우저를 만들 수 있고, socket 보다 더 손 쉽게 접근 가능

- 예전엔 파일(txt)에서 했던 것을 이제는 인터넷에서 바로 읽어올 수 있음

```python
import urllib.request, urllib.parse, urllib.error

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')
for line in fhand :
    print(line.decode().strip())
```

```python
import urllib.request, urllib.parse, urllib.error

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')

counts = dict()
for line in fhand :
    words = line.decode().split()
    for word in words :
        counts[word] = counts.get(word, 0) + 1
print(counts)
```

### BeautifulSoup
- BeautifulSoup을 활용한 웹 스크래핑
BeautifulSoup는 아주 강력한 라이브러리입니다. 지난 시간에 배운 urllib과 더불어 사용하면 다음과 같이 원하는 웹 페이지에 존재하는 모든 링크의 URL을 출력할 수 있습니다.

```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# Ignore SSL certificate errors SSL 인증서 오류를 무시해줌
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter - ')
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

tags = soup('a')   # a로 시작하는 태그를 크롤링하라는 의미
for tag in tags:   # a로 시작하는 태그 속에서 href가 붙는 것을 출력해라
    print(tag.get('href', None))
```

```python
import socket
mysocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysocket.connect(('data.pr4e.org', 80))
cmd = 'GET http://data.pr4e.org/intro-short.txt HTTP/1.0\r\n\r\n'.encode()
# 위에서 마지막에 encode를 쓴 것은 서버쪽이 알아들수 있도록 해석해주는 의미
mysocket.send(cmd)

while True :
    data = mysocket.recv(512)  # recv(숫자) 는 데이터를 불러오는 속도를 의미하는데, 너무 크면 시스템이 버벅되서 그른 듯
    if (len(data) < 1 ) :
        break
    print(data.decode(),end='')  # 다시 내 컴퓨터에서 볼 거니까 서버의 URF-8을 유니코드로 decode해서 출력

mysocket.close()
```
